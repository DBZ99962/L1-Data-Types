#### 1. Write a program to take principal amount, rate and interest as input and calculate simple interest.

```python
# Program to calculate simple interest
p = float(input("Enter principal amount: "))
t = float(input("Enter time period (years): "))
r = float(input("Enter rate of interest: "))

si = (p * t * r) / 100
print(f"Principal: {p}")
print(f"Time Period: {t}")
print(f"Rate of Interest: {r}")
print(f"Simple Interest: {si}")
print(f"Total Amount: {p + si}")
```

#### 3. Write a program to take a list of numbers as input (in a single line) and print the sum, avg, max and min elements.

```python
# Program to find sum, average, max and min of list numbers
numbers_input = input("Enter numbers separated by spaces or commas: ")
# Replace commas with spaces to handle both formats
numbers_input = numbers_input.replace(',', ' ')
numbers = list(map(float, numbers_input.split()))

total_sum = sum(numbers)
average = total_sum / len(numbers)
max_num = max(numbers)
min_num = min(numbers)

print(f"Numbers: {numbers}")
print(f"Sum: {total_sum}")
print(f"Average: {average}")
print(f"Maximum: {max_num}")
print(f"Minimum: {min_num}")
```

#### 5. Write a program to take a list as input containing results of 10 coin tosses as 'H' or 'T'. Jon gets a point for each 'T' and Tom gets a point for each 'H'. print the points recieved by both.

```python
# Program to count points from coin tosses
tosses_input = input("Enter 10 coin tosses (H or T separated by spaces): ")
tosses = tosses_input.split()

jon_points = 0  # Jon gets a point for each 'T'
tom_points = 0  # Tom gets a point for each 'H'

for toss in tosses:
    if toss.upper() == 'T':
        jon_points += 1
    elif toss.upper() == 'H':
        tom_points += 1

print(f"Toss results: {tosses}")
print(f"Jon's points (T count): {jon_points}")
print(f"Tom's points (H count): {tom_points}")
if jon_points > tom_points:
    print(f"Jon wins by {jon_points - tom_points} points!")
elif tom_points > jon_points:
    print(f"Tom wins by {tom_points - jon_points} points!")
else:
    print("It's a tie!")



  

```

#### 7. Take a list as input in a single line containing 10 numbers as marks out of 100. The even positions in the list are marks of Tom and the odd positions are the marks of Jon. Calculate the average marks of both.

```python
# Program to calculate average marks for Tom (even positions) and Jon (odd positions)
marks_input = input("Enter 10 marks separated by spaces: ")
marks = list(map(float, marks_input.split()))

# Even positions (1-indexed): positions 2, 4, 6, 8, 10 -> indices 1, 3, 5, 7, 9
tom_marks = [marks[i] for i in range(1, len(marks), 2)]

# Odd positions (1-indexed): positions 1, 3, 5, 7, 9 -> indices 0, 2, 4, 6, 8
jon_marks = [marks[i] for i in range(0, len(marks), 2)]

tom_avg = sum(tom_marks) / len(tom_marks)
jon_avg = sum(jon_marks) / len(jon_marks)

print(f"All marks: {marks}")
print(f"\nTom's marks (even positions): {tom_marks}")
print(f"Tom's average: {tom_avg}")
print(f"\nJon's marks (odd positions): {jon_marks}")
print(f"Jon's average: {jon_avg}")



    


```

#### 9. Create a dictionary having three items. Key of each item is the name of a student, value associated with each key is a list of marks in five subjects. Given this dictionary print the average score of each student.

```python
# Program to create dictionary with student names and marks, and print average
student_marks = {}

print("Enter marks for 3 students (5 subjects each):")
for i in range(3):
    name = input(f"Enter name of student {i+1}: ")
    marks_input = input(f"Enter marks for {name} (5 subjects separated by spaces): ")
    marks = list(map(float, marks_input.split()))
    student_marks[name] = marks

print("\n" + "="*50)
print("Student Averages:")
print("="*50)

for name, marks in student_marks.items():
    average = sum(marks) / len(marks)
    print(f"\n{name}:")
    print(f"  Marks: {marks}")
    print(f"  Average Score: {average:.2f}")
```

#### 10. Take a string as input(may contain repetitive words), print a list consisting of unique words from the given string.

```python
# Program to extract unique words from a string
string = input("Enter a string (may contain repetitive words): ")

words = string.split()
unique_words = list(set(words))
unique_words_ordered = []

# Maintain order of first appearance
for word in words:
    if word not in unique_words_ordered:
        unique_words_ordered.append(word)

print(f"\nOriginal string: {string}")
print(f"Words: {words}")
print(f"Unique words (ordered): {unique_words_ordered}")
print(f"Number of unique words: {len(unique_words_ordered)}")
```
